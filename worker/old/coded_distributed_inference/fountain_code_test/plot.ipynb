```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
kn = np.asarray([(2, 4), (3, 4), (4, 7), (5, 7), (6, 8), (7, 10), (8, 10), (9, 12), (10, 14)])
ks = kn[:, 0]
ns = kn[:, 1]
rs = ns - ks

plt.plot(ks, ns, label='Block number for decoding')
plt.plot(ks, rs, label='redundancy')
plt.xlabel('k: source number of blocks')
plt.ylabel('n or r')
plt.legend()
plt.show()
```
